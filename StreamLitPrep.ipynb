```python
all_info = all_info.drop(columns=['w2vVector','business_status'])
```

```python
df_similar_columns = ['rating_x','user_ratings_total_x','num_restaurants','rest_mean_rating',
                      'num_bars', 'bars_mean_rating', 'num_art_galleries', 'art_gallery_mean_rating', 'num_gyms', 
                      'gyms_mean_rating', 'num_shopping', 'shopping_mean_rating', 'num_tourist_attract',
                      'tourist_mean_rating', 'num_casinos','casinos_mean_rating', 'num_supermarkets', 'supermarkets_mean_rating']
```

```python
df_similar_trial = all_info[df_similar_columns]
```

```python
df_similar_trial
```

Fix Null Values:

```python
df_similar_trial.isnull().sum()
```

```python
no_nulls = df_similar_trial.fillna(df_similar_trial.mean())
no_nulls.isnull().sum()
```

```python
no_nulls = no_nulls.drop(columns=['casinos_mean_rating'])
no_nulls.isnull().sum()
```

```python
no_nulls.head()
```

```python
no_nulls.to_csv(r'numerical_cols.csv',index=True, header=True, index_label=False)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(no_nulls)
```

```python
scaled_df = pd.DataFrame(scaled)
scaled_df
```

```python
similarity_matrix = pd.DataFrame(cosine_similarity(scaled_df))
```
